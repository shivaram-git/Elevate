```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Elevate\\Titanic-Dataset.csv")
df=pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head(10)
```

```python
print("Before Number of Null values:\n",df.isnull().sum())
```

```python
# Fill Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop Cabin (too many missing)
df.drop(columns=['Cabin'], inplace=True)

# Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
print("After Handling Number of Null values:\n",df.isnull().sum())
```

```python
no_of_duplicated_values=df.duplicated().sum()
print('Number of dupliicate values:',no_of_duplicated_values)
```

```python



df.describe(include='all')

# Value counts for categorical columns
print(df['Sex'].value_counts())

df['Embarked'].value_counts()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Class")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare Colored by Survival")
plt.show()
```

```python

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']].dropna(), hue='Survived')
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```
